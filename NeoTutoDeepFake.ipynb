```
#@title 3) Déclaration de fonctions Communes

import time

start_time = 0
sec = ''
def start():
  global start_time
  start_time = time.time()
def endtime():
  global sec
  elapsed_time = time.time() - start_time
  sec = "{0:.2f}".format(elapsed_time) + 's'
  
def download(item):
  start()
  src = goo_dir + '/' + item
  trg = wrk_dir + '/' + item
  cmd = 'cp -n ' + src + ' ' + trg
  !$cmd
  endtime()
  print('Download de ' + src + ' vers ' + trg + ' en ' + sec)

def download_and_unzip(zip, target):
  start()
  # 'download' (copy) depuis Google Drive
  goofile = goo_dir + '/' + zip
  zipfile = wrk_dir + '/' + zip
  !cp -n $goofile $zipfile
  # unzip des fichiers
  trg_dir = wrk_dir + '/' + target
  !unzip -jq $zipfile -d $trg_dir
  endtime()
  print('copie de ' + goofile + ' vers ' + zipfile + ' et extraction dans ' + trg_dir + ' en ' + sec)
  
def zip_and_upload(pattern, zip):
  start()
  # zip des fichiers
  srcitems = wrk_dir + '/' + pattern
  zipfile = wrk_dir + '/' + zip
  !zip -jq $zipfile $srcitems
  # 'upload' (copy) dans Google Drive
  goofile = goo_dir + '/' + zip
  !cp -n $zipfile $goofile
  endtime()
  print('zip de ' + srcitems + ' vers ' + zipfile + ' et copie dans ' + goofile + ' en ' + sec)
  
def upload(item):
  start()
  srcitem = wrk_dir + '/' + item
  # 'upload' (copy) dans Google Drive
  goofile = goo_dir + '/' + item
  !cp -n $zipfile $goofile
  endtime()
  print('zip de ' + srcitems + ' vers ' + zipfile + ' et copie dans ' + goofile + ' en ' + sec)
  
print('fonctions importées')
```
